```python
pca = PCA(n_components = 2)
X_train = pca.fit_transform(np.asarray(X_cancer_train.T))
X_test = pca.transform(np.asarray(X_cancer_test.T))
cancer_explained_variance = pca.explained_variance_ratio_
```

```python
cancer_explained_variance
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,np.asarray(y_cancer_train.T).ravel())
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(np.asarray(y_cancer_test.T).ravel(),y_pred)
```

```python
labels = ['breast','prostate','lung','collerectal',
                  'lymphoma','bladder','melanoma','uterus',
                  'leukemia','renal','pancreas','ovary',
                  'meso','cns']
cancer_keys = list(map(str,range(14)))
cancer_dict = dict(zip(cancer_keys,labels))
```

```python
cm = pd.DataFrame(cm)
cm['labels'] = labels
cm.index = cm['labels']
cm = cm.iloc[:, :-1]
cm.columns = labels
cm
```

```python
label_color_dict = {'breast':'red',
                    'prostate':'green',
                    'lung':'blue',
                    'collerectal':'coral',
                    'lymphoma':'ivory',
                    'bladder':'lavender',
                    'melanoma':'papayawhip',
                    'uterus':'aqua',
                    'leukemia':'b',
                    'renal':'seagreen',
                    'pancreas':'crimson',
                    'ovary':'deeppink',
                    'meso':'crimson',
                    'cns':'darkslategray'}

label_color_dict = {label:idx for idx,label in enumerate(np.unique(labels))}
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, np.asarray(y_cancer_test.T).ravel()
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green', 'blue','coral',
                                                  'ivory','lavender','papayawhip','aqua',
                                                  'b','seagreen','crimson','deeppink',
                                                  'crimson','darkslategray')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue','coral',
                                    'ivory','lavender','papayawhip','aqua',
                                    'b','seagreen','crimson','deeppink',
                                    'crimson','darkslategray'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

### Self Organizing Map (SOM)

```python
from sklearn.preprocessing import MinMaxScaler
from minisom import MiniSom
sc = MinMaxScaler(feature_range = (0,1))
```

```python
yeast_data.shape
```

### Datos de Levadura

```python
yeast_scaled = sc.fit_transform(yeast_data)
som = MiniSom(x = 10, y = 10, input_len = 17, sigma=1.0, learning_rate = 0.5)
som.random_weights_init(yeast_scaled)
som.train_random(data = yeast_scaled, num_iteration = 100)
```

```python
from pylab import bone, pcolor, colorbar, plot, show
bone()
pcolor(som.distance_map().T)
colorbar() 

for i, x in enumerate(yeast_scaled):
    w = som.winner(x)
    plot(w[0] + 0.5,
         w[1] + 0.5,
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2)
show()
```
